```python
# استيراد المكتبات
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# قراءة ملف البيانات
df = pd.read_csv('patient_data.csv')  # تأكدي إن الملف مرفوع واسم الملف صح

# تصفية المصابين بالسكري وتجميعهم حسب المنطقة
diabetes_by_region = df[df['Diabetes'] == 'Yes'].groupby('Region')['Name'].count()

# رسم Bar Chart
plt.figure(figsize=(8, 5))  # تحديد حجم الرسم
diabetes_by_region.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Number of diabetic patients by region", fontsize=14)
plt.xlabel("Region", fontsize=12)
plt.ylabel("Number of Patients", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Diabetes Cases by Region (Simple Analysis)

The bar chart shows the number of diabetes cases in each region. According to the data, both **Riyadh** and **Jeddah** have one case each. Since the dataset is small, it's not possible to identify strong patterns or trends, but this chart gives an initial look at how diabetes is distributed between regions.

```python
plt.figure(figsize=(7, 5))
sns.boxplot(data=df, x='Gender', y='BMI', palette='pastel')
plt.title("Body Mass Index (BMI) Distribution by gender", fontsize=14)
plt.xlabel("Gender", fontsize=12)
plt.ylabel("BMI", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

**BMI** Distribution by Gender (Simple Analysis)

The data shows that male patients have a higher average Body Mass Index (**BMI**) compared to female patients. There is also more variation in BMI among males. This may suggest that males in this dataset are more likely to be overweight, which could be linked to higher risk of health issues like diabetes.
